```python
import numpy as np
import matplotlib.pyplot as plt

v=int(input("Enter velocity of the object: "))
t=10
dt=1
x=0
F=int(input("Enter force applied to the object: "))
m=int(input("Enter mass of the object: "))
a=F/m


positions = []
time = []
velocities = []
acceleration =[]

for i in range(0, t+1, dt):
  positions.append(x)
  time.append(i)
  velocities.append(v)
  acceleration.append(a)

  x=x+(v*dt)
  v=v+(a*dt)

print(positions)
print(time)
print(velocities)
print(acceleration)

plt.plot(time, positions)
plt.xlabel("Time")
plt.ylabel("Position")
plt.title("Newton's Second Law: Time vs Position")
plt.show()

plt.plot(time, velocities)
plt.xlabel("Time")
plt.ylabel("Velocity")
plt.title("Newton's Second Law: Time vs Velocity")
plt.show()

plt.plot(time, acceleration)
plt.xlabel("Time")
plt.ylabel("Acceleration")
plt.title("Newton's Second Law: Time vs Acceleration")
plt.show()
```

problem statement
you have a particle of mas m moving in one dimension (the x-axis 1D)
a constant force F acts on the particle in x direction
time step dt=0.1s
total simulation time t=10s

Using Newtons Second law (F=ma)
Newton's Second Law of motion states that an objects accelaration is directly proportional to its mass, often summarixed as the equation F=ma(mass*acceleration)
